```python
# Import dependencies
import pandas as pd
import gmaps
import requests

# Import the API key.
from config import g_key
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Store the CSV created in part one into a DataFrame.
city_data_df = pd.read_csv("data/WeatherPy_database.csv")
city_data_df.head()
```

```python
city_data_df.dtypes
```

```python
#See if there are any cities that have rain.
raining_cities = city_data_df.loc[city_data_df["Rain (Inches)"] > 0]
raining_cities.head()
```

```python
#See if there are any cities that have snow.
snowing_cities = city_data_df.loc[city_data_df["Snow (Inches)"] > 0]
snowing_cities.head()
```

```python
#Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
```

```python
#ASk the customer if they want it to be raining
rain_amount = input("Do you want it to be raining? (yes/no) ")
```

```python
#Ask the customer if they want it to be snowing
snow_amount = input("Do you want it to be snowing (yes/no) ")
```

```python
if rain_amount == "no" and snow_amount == "no":
    filtered_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & 
                                          (city_data_df["Max Temp"] >= min_temp) &
                                          (city_data_df["Rain (Inches)"] == 0) &
                                          (city_data_df["Snow (Inches)"]==0)]
elif rain_amount == "yes" and snow_amount == "no":
    filtered_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & 
                                          (city_data_df["Max Temp"] >= min_temp) &
                                          (city_data_df["Rain (Inches)"] > 0.0) &
                                          (city_data_df["Snow (Inches)"]==0)]
elif rain_amount == "no" and snow_amount == "yes":
    filtered_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & 
                                          (city_data_df["Max Temp"] >= min_temp) &
                                          (city_data_df["Rain (Inches)"] == 0) &
                                          (city_data_df["Snow (Inches)"]>0.0)]
else:
    filtered_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & 
                                          (city_data_df["Max Temp"] >= min_temp) &
                                          (city_data_df["Rain (Inches)"] > 0.0) &
                                          (city_data_df["Snow (Inches)"]>0.0)]
```

```python
#Examine the first rows of the DataFrame.
filtered_cities_df.head()
```

```python
#Get a count of the number of viable cities for the inputs.
filtered_cities_df.count()
```

```python
#Remove any "n/a" rows.
filtered_cities_df = filtered_cities_df.dropna()
filtered_cities_df.count()
```

```python
#Create a hotels DataFrame containing the city, country, max temperature, latitude, longitude, and current weather.
hotel_df = filtered_cities_df[["City","Country","Max Temp","Lat","Lng","Current Description"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
# Set the parameters to search for a hotel.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

#Iterate through the DataFrame
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]
    
    #Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"
    
    #Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    #Make request and get JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    #Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index,"Hotel Name"] = hotels["results"][0]["name"]
    except(IndexError):
        print("Hotel not found... skipping.")
    
```

```python
#Create an output file (CSV).
output_data_file = "data/WeatherPy_vacation.csv"
hotel_df.to_csv(output_data_file, index_label="City_ID")
```

```python
#Create a map of the locations of the selected hotels.
locations = hotel_df[["Lat","Lng"]]
fig = gmaps.figure(center=(30.0,31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations)
fig.add_layer(marker_layer)

#Call the figure to plot the data.
fig
```

```python
#Create an information box to add to the map.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
#Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
#Add marker labels to the markers on the map.
locations = hotel_df[["Lat","Lng"]]
fig = gmaps.figure(center=(30.0,31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
#Call the figure to plot the data.
fig
```
